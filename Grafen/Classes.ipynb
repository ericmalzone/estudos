```python
class Carro:
    
    def __init__(self, marca, ano, cor):
        self.marca = marca
        self.ano = ano
        self.cor = cor
        
    def imprimir(self, texto):
        print("A mensagem é: {}".format(texto)) 
        
    def setMarca(self, marca):
        self.marca = marca
        return self.marca
```

```python
class Jogadores:
    
    def __init__(self):
        self.arquivo = "jogadores.txt"
        #pass
    
    def criarJogadores(self):
        arquivoJ = open(self.arquivo, "w")
        arquivoJ.write("Lionel Messi,Atacante,Barcelona,Esquerda\n")
        arquivoJ.write("Philippe Coutinho,Meia,Barcelona,Direita\n")
        arquivoJ.write("Bruno Peres,Zagueiro,São Paulo,Direita\n")
        arquivoJ.write("Hernanes,Meia,São Paulo,Esquerda\n")
        arquivoJ.write("Cristiano Ronaldo,Atacante,Juventus,Direita\n")
        arquivoJ.write("Leonardo Bonucci,Atacante,Juventus,Direita\n")
        arquivoJ.write("Orinho,Lateral Direito,Santos,Direita\n")
        arquivoJ.write("Jorge,Lateral Esquerdo,Santos,Direita\n")
        arquivoJ.write("Lucas Verissimo,Zagueiro,Santos,Esquerda\n")
        arquivoJ.write("Fagner,Zagueiro,Corinthians,Direita\n")
        arquivoJ.write("Cassio,Goleiro,Corinthians,Esquerda\n")
        arquivoJ.write("Rogerio Ceni,Goleiro,São Paulo,Direita")
        arquivoJ.close()
    
    def ler_arquivo(self):
        self.arqJogadores = open(self.arquivo, "r")
        self.jogadores = self.arqJogadores.read()
        return self.jogadores
    
    def get_posicoes(self, lst_jogadores):
        self.atacante = []
        self.meia = []
        self.zagueiro = []
        self.lateralDireito = []
        self.lateralEsquerdo = []
        self.goleiro = []
        self.jogadores = lst_jogadores
        for jogador in self.jogadores:
            if (jogador["Posicao"] == "Atacante"):
                self.atacante.append(jogador)
            elif (jogador["Posicao"] == "Meia"):
                self.meia.append(jogador)
            elif (jogador["Posicao"] == "Zagueiro"):
                self.zagueiro.append(jogador)
            elif (jogador["Posicao"] == "Lateral Direito"):
                self.lateralDireito.append(jogador)
            elif (jogador["Posicao"] == "Lateral Esquerdo"):
                self.lateralEsquerdo.append(jogador)
            elif (jogador["Posicao"] == "Goleiro"):
                self.goleiro.append(jogador) 
        print("Jogadores Atacantes: {}".format(self.atacante))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores Meias: {}".format(self.meia))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores Zagueiros: {}".format(self.zagueiro))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores Laterais Direitos: {}".format(self.lateralDireito))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores Laterais Esquerdos: {}".format(self.lateralEsquerdo))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores Goleiros: {}".format(self.goleiro))
        
    def get_times(self, lst_jogadores):
        self.sao_paulo = []
        self.bercelona = []
        self.juventus = []
        self.santos = []
        self.corinthians = []
        self.jogadores = lst_jogadores
        for jogador in self.jogadores:
            if (jogador["Time"] == "São Paulo"):
                self.sao_paulo.append(jogador)
            elif (jogador["Time"] == "Barcelona"):
                self.bercelona.append(jogador)
            elif (jogador["Time"] == "Juventus"):
                self.juventus.append(jogador)
            elif (jogador["Time"] == "Santos"):
                self.santos.append(jogador)
            elif (jogador["Time"] == "Corinthians"):
                self.corinthians.append(jogador)
        print("Jogadores do São Paulo: {}".format(self.sao_paulo))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores do Barcelona: {}".format(self.bercelona))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores do Juventus: {}".format(self.juventus))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores do Santos: {}".format(self.santos))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores do Corinthians: {}".format(self.corinthians))        
        
        
    def get_lado_chute(self, lst_jogadores):
        self.esquerda = []
        self.direita = []
        self.jogadores = lst_jogadores
        for jogador in self.jogadores:
            if (jogador["Perna"] == "Direita"):
                self.direita.append(jogador)
            else:
                self.esquerda.append(jogador)
        print("Jogadores Destros: {}".format(self.direita))
        print("-------------------------------------------------------------------------------------------------------------")
        print("-------------------------------------------------------------------------------------------------------------")
        print("Jogadores Canhotos: {}".format(self.esquerda))
```

```python
class Jogadores:
    
    def __init__(self):
        #self.nome = _nome
        #self.posicao  = _posicao
        #self.time = _time
        #self.perna = _perna
        self.arquivo = None
        self.lstJogadores = []
        
    def criar_arquivo(self):
        arquivoJ = open("jogadores2.txt", "w")
        arquivoJ.write("Lionel Messi,Atacante,Barcelona,Esquerda\n")
        arquivoJ.write("Philippe Coutinho,Meia,Barcelona,Direita\n")
        arquivoJ.write("Bruno Peres,Zagueiro,São Paulo,Direita\n")
        arquivoJ.write("Hernanes,Meia,São Paulo,Esquerda\n")
        arquivoJ.write("Cristiano Ronaldo,Atacante,Juventus,Direita\n")
        arquivoJ.write("Leonardo Bonucci,Atacante,Juventus,Direita\n")
        arquivoJ.write("Orinho,Lateral Direito,Santos,Direita\n")
        arquivoJ.write("Jorge,Lateral Esquerdo,Santos,Direita\n")
        arquivoJ.write("Lucas Verissimo,Zagueiro,Santos,Esquerda\n")
        arquivoJ.write("Fagner,Zagueiro,Corinthians,Direita\n")
        arquivoJ.write("Cassio,Goleiro,Corinthians,Esquerda\n")
        arquivoJ.write("Rogerio Ceni,Goleiro,São Paulo,Direita\n")
        arquivoJ.write("Eric Malzone,Atacante,São Paulo,Direita")
        arquivoJ.close()
        
    def ler_arquivo(self):    
        self.arquivo = open("jogadores2.txt", "r")
        
    def fechar_arquivo(self):
        self.arquivo.close()
        
    def ler_jogadores(self):
        for linha in self.arquivo.readlines():
            palavras = linha.strip("\n").split(",")
            dic = {}
            dic["Nome"] = palavras[0]
            dic["Posicao"] = palavras[1]
            dic["Time"] = palavras[2]
            dic["Perna"] = palavras[3]
            self.lstJogadores.append(dic)
    
    def get_posicoes(self):
        for jogador in self.lstJogadores:
            if jogador["Posicao"] == "Goleiro":
                print("Jogador: {} é goleiro!".format(jogador["Nome"]))
            elif jogador["Posicao"] == "Atacante":    
                print("Jogador: {} é atacante!".format(jogador["Nome"]))
            elif jogador["Posicao"] == "Zagueiro":    
                print("Jogador: {} é zagueiro!".format(jogador["Nome"]))    
```
